**6) FINDING THE OUTLINERS AND REPLACING THEM**

For `AGE`

```python
sns.boxplot(df['Age'])
```

```python
q1 = df['Age'].quantile(0.25)
q3 = df['Age'].quantile(0.75)
iqr = q3 - q1
iqr
```

```python
df.loc[df['Age'] < q1 - 1.5*iqr, 'Age'] = q1
df.loc[df['Age'] > q3 + 1.5*iqr, 'Age'] = q3
sns.boxplot(x=df['Age'])
```

REMOVED OUTLINERS

FOR `CREDIT SCORE`

```python
sns.boxplot(df['CreditScore'])
```

```python
Q1=df['CreditScore'].quantile(0.25)
Q3=df['CreditScore'].quantile(0.75)
IQR=Q3-Q1
whisker_width = 2.0
credscore_outliers = df[(df['CreditScore'] < Q1 - whisker_width*IQR) | (df['CreditScore'] > Q3 + whisker_width*IQR)]
credscore_outliers.count()
```

```python
lower_whisker = Q1 -(whisker_width*IQR)
upper_whisker = Q3 + (whisker_width*IQR)
df['CreditScore']=np.where(df['CreditScore']>upper_whisker,upper_whisker,np.where(df['CreditScore']<lower_whisker,lower_whisker,df['CreditScore']))
```

```python
sns.boxplot(df['CreditScore'])
```

**8) SPLITTING DEPENDENT AND INDEPENDENT VARIABLES**

```python
X = df.iloc[:,3:-1]
```

```python
X.head()
```

```python
y = df.iloc[:,-1]
```

**9) SCALING THE INDEPENDENT VARIABLES**

```python
from sklearn.preprocessing import StandardScaler
scale = StandardScaler()
X = scale.fit_transform(X)
X
```

**10) SPLITING THE DATA INTO TEST AND TRAINING**

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y ,random_state=1, shuffle=True, train_size = 0.75)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn import linear_model
reg = linear_model.LogisticRegression()
reg.fit(X_train, y_train)
```

```python
y_pred = reg.predict(X_test)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```
